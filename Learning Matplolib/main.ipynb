```python
#Importation de matplot
import matplotlib

#Version de matplot
print(matplotlib.__version__)

#Importation de pyplot
import matplotlib.pyplot as plt
```

```python
#Représentation d'un point dans un graphique
import numpy as np
xpoints = np.array([1])
ypoints = np.array([3])
plt.plot(xpoints,ypoints,"*")
plt.show()
```

```python
#Relier deux points
xpoints = np.array([1,5])
ypoints = np.array([3,8])

font1 = {'family':'arial','color':'blue','size':10}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Est-ce que bière est doux ?",fontdict=font1,loc="right")
plt.xlabel("Bière")
plt.ylabel("Doux")
plt.plot(xpoints,ypoints,marker="o")
plt.show()
plt.ion()
```

```python

#Diagramme circulaire
pie = plt.pie(ypoints,labels=ypoints)
plt.legend(title = "C'est doux oh")
plt.show()
plt.ion()

```
